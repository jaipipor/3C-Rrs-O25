# 01 — Quick Start: demo of the 3C Rrs model

This notebook shows a minimal, runnable example to:

1. add the repository `src/` to `sys.path` (so no `pip install -e .` is required),
2. load a single example spectrum, instantiate the model, run a fit, and plot results.

You can run the cells top-to-bottom. The notebook is written to be safe for users
who cloned the repo and installed dependencies in a virtual environment.

```python
# Robust helper: find repo `src/rrs3c` by searching upward and insert it into sys.path.
# Run this cell first in the notebook before importing rrs3c.
# Imports
import sys
import numpy as np
import lmfit as lm
import matplotlib.pyplot as plt
from pathlib import Path
from rrs3c.model import rrs_model_3C
print('Successfully imported rrs_model_3C')

def find_src_rrs3c(start_dir=Path.cwd(), max_parents=8):
    p = start_dir.resolve()
    for _ in range(max_parents + 1):
        candidate = p / "src" / "rrs3c"
        if candidate.exists() and (candidate / "model.py").exists():
            return (p / "src").resolve()
        p = p.parent
    return None

src_dir = find_src_rrs3c()
print("Notebook working dir:", Path.cwd())
if src_dir is None:
    # Print helpful diagnostic listing to help you locate the repo
    print("\nCould not find 'src/rrs3c' searching up from the notebook cwd.")
    # show immediate directory contents (small amount)
    try:
        print("\nContents of cwd:")
        for i, it in enumerate(sorted(Path.cwd().iterdir())):
            if i < 50:
                print(" ", it.name)
            else:
                print("  ...")
                break
    except Exception as e:
        print("  (error listing cwd)", e)
    # Also print a suggestion
    raise RuntimeError(
        "Please run the notebook from the repository root, or set PYTHONPATH to point to the repo 'src' folder.\n"
        "Example (PowerShell): $env:PYTHONPATH = \"$PWD\\src\" ; jupyter lab\n"
        "Or update this cell to point to the correct src folder explicitly."
    )

# Insert src_dir at front of sys.path so imports find local package first
sys.path.insert(0, str(src_dir))
print("Inserted into sys.path[0]:", sys.path[0])

# show the rrs3c path if importable (will raise if still not found)
try:
    import importlib
    spec = importlib.util.find_spec("rrs3c")
    if spec is not None:
        print("rrs3c package discovered at:", spec.origin or spec.submodule_search_locations)
    else:
        print("rrs3c not discoverable by importlib (odd).")
except Exception as e:
    print("importlib check failed:", e)
```

## Load an example spectrum

The example CSV used in the repository is expected at `examples/example_single_spectrum.csv`.
We read the CSV directly if present and fall back to a small synthetic spectrum.

```python
# locate example file (if present) and build a simple spectrum
ex = Path(Path.cwd()) / ".." / "examples" / "singlespectrum" / "data" / "example_single_spectrum.csv"
print(ex.exists())
if ex.exists():
    import pandas as pd
    df = pd.read_csv(ex, index_col=0, skiprows=15)
    wl = df.index.values.astype(float)
    # pick first two columns as Li and Lt if present, and Es if present in 3rd column
    cols = df.columns.tolist()
    Li = df.iloc[:,0].values if df.shape[1] >= 1 else np.zeros_like(wl)
    Lt = df.iloc[:,1].values if df.shape[1] >= 2 else np.zeros_like(wl)
    Es = df.iloc[:,2].values if df.shape[1] >= 3 else np.ones_like(wl)
    print('Loaded example CSV with', wl.size, 'wavelengths')
else:
    # build synthetic wl and simple spectra for demonstration
    wl = np.arange(350, 921)
    Es = 1e3 * np.exp(-0.002*(wl-440))  # synthetic shape (arbitrary)
    Li = 0.01 * Es
    Lt = 0.02 * Es
    print('Using synthetic demo spectrum with', wl.size, 'wavelengths')
```

## Instantiate the model and run a single quick fit

Call the model and fit Lt/Es to the model given its sun-view geometry.

The parameter bounds and initial estimates can be tuned to the local environment and geometric conditions.

```python
# Instantiate model
model = rrs_model_3C()

# Build simple parameter object with lmfit (this mirrors examples in the repo)
params = lm.Parameters()
params.add_many(
    ('C', 5, True, 0.1, 50, None),
    ('N', 1, True, 0.01, 100, None),
    ('Y', 0.5, True, 0.01, 5, None),
    ('SNAP', 0.015, True, 0.005, 0.03, None),
    ('Sg', 0.015, True, 0.005, 0.03, None),
    ('rho', 0.02, False, 0, 0.03, None),
    ('rho_d', 0.0, True, 0, 10, None),
    ('rho_s', 0.0, True, -.01, .01, None),
    ('alpha', 0.2, True, 0, 2, None),
    ('beta', 0.05, True, 0, 1, None)
)

weights = np.where((wl >= 760) & (wl <= 765), 0,
          np.where(wl < 400, 1,
          np.where(wl > 800, 5, 1)))

geom=(59, 35, 100)  #  sun-view geometry for the example case
anc = (4, 60, 1013.25)

# Li/Es and Lt/Es are used by the model-to-data-fit
try:
    out, Rrs_mod, Rg = model.fit_LtEs(wl, Li/Es, Lt/Es, params, weights, geom, anc)
    print('Fit completed; use `Rrs_mod` and `Rg` for plotting')
except Exception as e:
    print('Fit raised an exception (this can happen if inputs are synthetic).')
    print('Exception:', e)
```

## Plot results

If the fit succeeded two arrays are available: `Rrs_mod` (modeled water-leaving reflectance) and `Rg` (surface-reflected). We plot measured Lt/Es and model output.

```python
try:
    plt.figure(figsize=(9,4))
    plt.plot(wl, Lt/Es, label='Lt/Es measured')
    plt.plot(wl, (Rrs_mod + Rg), '--', label='Lt/Es modelled')
    plt.plot(wl, Rrs_mod, '--', label='Rrs modelled')
    plt.plot(wl, Rg, label='R_g: glint signal')
    plt.plot(wl, Lt/Es - Rg, label='Rrs: 3C estimate')
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Reflectances (sr^-1)')
    plt.legend()
    plt.grid(True)
    plt.show()
except NameError:
    print('No results to plot (fit did not produce Rrs_mod/Rg).')
```
