# 02 — Time-Series Processing

Process a time-series of spectra and produce a NetCDF with Es, modeled Rrs and Rg.
This notebook mirrors the `examples/run_timeseries.py` wrapper but in an interactive form.

```python
# ruff: noqa: E402
import sys
import numpy as np
import lmfit as lm
import matplotlib.pyplot as plt
from pathlib import Path
import time
from datetime import datetime
import pandas as pd
import xarray as xr

def find_src_rrs3c(start_dir=Path.cwd(), max_parents=8):
    p = start_dir.resolve()
    for _ in range(max_parents + 1):
        candidate = p / "src" / "rrs3c"
        if candidate.exists() and (candidate / "model.py").exists():
            return (p / "src").resolve()
        p = p.parent
    return None

src_dir = find_src_rrs3c()
print("Notebook working dir:", Path.cwd())

if src_dir is None:
    raise RuntimeError("Could not find src/rrs3c. Make sure the notebook is inside the repository.")

repo_root = src_dir.parent
sys.path.insert(0, str(src_dir))
sys.path.insert(0, str(repo_root))
print("Repository root added to path:", repo_root)

from rrs3c.model import rrs_model_3C_O25
print('Successfully imported rrs_model_3C_O25')
```

## Load raw Jetty data (example CSV)
If your CSV is in `examples/example_time_series_data.csv` the loader below will read it. Otherwise adjust the path.

```python
from examples.timeseries.src.utils import load_jetty_data, solar_az_el, flags_jetty

# Use repo_root from the previous cell (if you want to keep it in scope, you may need to make it global or recompute)
# For simplicity, recompute repo_root:
repo_root = Path(find_src_rrs3c()).parent
infile = repo_root / "examples" / "timeseries" / "data" / 'example_time_series_data.csv'

if not infile.exists():
    raise FileNotFoundError(f"Data file not found at {infile}. Please place it there or adjust the path.")

instrument, variable, times, l1, l_end, delta_l, spec = load_jetty_data(str(infile))
print('Loaded raw:', spec.shape, 'spectra')
```

## Convert times to pandas datetime and factorize into unique slots

```python
times_pd = pd.to_datetime(times, format='%Y%m%d %H%M%S')
idx_inv, uniques = pd.factorize(times_pd)
unique_times = uniques.to_pydatetime()
anc = (4, 60, 1013.25)
N = len(unique_times)
wl = np.arange(350, 921)
nl = len(wl)
```

## Pre-allocate matrices for Es, Li, Lt, Rrs, etc.

```python
Es_mat = np.full((N, nl), np.nan)
Li_east = np.full((N, nl), np.nan)
Lt_east = np.full((N, nl), np.nan)
Rrs_east = np.full((N, nl), np.nan)
Rrs_mod_east = np.full((N, nl), np.nan)
Rg_east = np.full((N, nl), np.nan)

Li_west = np.full((N, nl), np.nan)
Lt_west = np.full((N, nl), np.nan)
Rrs_west = np.full((N, nl), np.nan)
Rrs_mod_west = np.full((N, nl), np.nan)
Rg_west = np.full((N, nl), np.nan)
```

## Diagnostics and storage for fit objects (initialized before the loop)

```python
idx550 = np.argmin(np.abs(wl - 550))
resid_rmse_east = np.full(N, np.nan)
resid_rmse_west = np.full(N, np.nan)
resid550_east = np.full(N, np.nan)
resid550_west = np.full(N, np.nan)
rho_east = np.full(N, np.nan)
rho_d_east = np.full(N, np.nan)
rho_s_east = np.full(N, np.nan)
rho_west = np.full(N, np.nan)
rho_d_west = np.full(N, np.nan)
rho_s_west = np.full(N, np.nan)
Rg550_east = np.full(N, np.nan)
Rg550_west = np.full(N, np.nan)

# Lists to keep whole fit Result objects for later inspection
res_east_list = [None] * N
res_west_list = [None] * N
```

# Build 3C model instance
## Model is instantiated outside the loop

```python
model = rrs_model_3C_O25()  # use default data_folder or override as needed
```

## Loop

```python
start_time = time.time()
# Loop over unique time bins
for i, t0 in enumerate(unique_times):
    mask = idx_inv == i
    vars_i = variable[mask]

    # select Es tag
    if 'ES' in vars_i:
        tag_es = 'ES'
    elif 'ES-VIS' in vars_i:
        tag_es = 'ES-VIS'
    else:
        continue

    # Extract Es spectrum and interpolate to global wl
    idx_es = np.where((variable == tag_es) & mask)[0]
    l_start, l_stop = float(l1[idx_es[0]]), float(l_end[idx_es[0]])
    raw = spec[idx_es, :int(l_stop - l_start + 1)][0]
    # build local wavelength axis for this spectrum
    wl_row = np.linspace(l_start, l_stop, raw.size)
    # interpolate onto global wl grid
    Es_interp = np.interp(wl, wl_row, raw)
    Es_mat[i, :] = Es_interp

    # Flag dark/red
    dark, red, anom= flags_jetty(wl, Es_interp)
    # Compute solar position
    Az, El = solar_az_el(t0, lat=53.001788, lon=4.789151)

    if not dark and not red and El >= 5:
        # process east and west
        for side in ('EAST', 'WEST'):
            sky_tag = f'LSKY-{side}'
            surf_tag = f'LSFC-{side}'
            if sky_tag in vars_i and surf_tag in vars_i:
                # extract raw Li, Lt
                idx_ls = np.where((variable == sky_tag) & mask)[0]
                idx_lu = np.where((variable == surf_tag) & mask)[0]
                raw_Li = spec[idx_ls, :int(l_stop - l_start + 1)][0]
                raw_Lt = spec[idx_lu, :int(l_stop - l_start + 1)][0]

                # build local wavelength axis and interpolate to global wl
                wl_row = np.linspace(l_start, l_stop, raw_Li.size)
                Li_interp = np.interp(wl, wl_row, raw_Li)
                Lt_interp = np.interp(wl, wl_row, raw_Lt)

                # store interpolated spectra
                # fit 3C model
                params = lm.Parameters()
                
                # O25
                params.add_many(
                    ('C', 5, True, 0.1, 50, None),
                    ('N', 1, True, 0.01, 100, None),
                    ('Y', 0.5, True, 0.01, 5, None),
                    ('SNAP', 0.015, True, 0.005, 0.03, None),
                    ('Sg', 0.015, True, 0.005, 0.03, None),
                    ('rho', 0.02, True, 0, 0.03, None),
                    ('rho_d', 0.0, True, 0, 10, None),
                    ('rho_s', 0.01, True, -.1, .1, None),
                    ('delta', 0.0, False, -0.01, 0.01, None),
                    ('alpha', 0.2, True, 0, 2, None),
                    ('beta', 0.05, True, 0, 1, None)
                )
                
                

                weights = np.where((wl >= 750) & (wl <= 775), 0, # Remove the ugly H2O feature
                    np.where(wl < 400, 1, # Some potential troubles at the UV
                    np.where(wl > 800, 5, 1))) # Let's power-up the NIR


                geom = (90 - El, 35, abs(Az - (135 if side == 'EAST' else 225)))                
                # call fit_Ltes with interpolated spectra
                res, Rrs_mod, Rg = model.fit_LtEs(
                    wl, Li_interp/Es_interp, Lt_interp/Es_interp, 
                    params, weights, geom, anc, verbose=False,
                )
                LtEs_obs = Lt_interp/Es_interp
                LtEs_mod = Rrs_mod + Rg

                # Compute residual in Lt/Es space as requested by the user
                # resid_vec = LtEs_mod - LtEs_obs
                resid_vec = LtEs_mod - LtEs_obs
                rmse_val = np.sqrt(np.nanmean((resid_vec)**2))
                resid550_val = resid_vec[idx550]

                # Observed and modelled Rrs (Lt/Es - Rg)
                Rrs_obs = LtEs_obs - Rg
                
                # Store into pre-allocated arrays and diagnostics
                if side == 'EAST':
                    Li_east[i, :] = Li_interp
                    Lt_east[i, :] = Lt_interp
                    Rrs_east[i, :] = Rrs_obs
                    Rrs_mod_east[i, :] = Rrs_mod
                    Rg_east[i, :] = Rg

                    resid_rmse_east[i] = rmse_val
                    resid550_east[i] = resid550_val
                    Rg550_east[i] = Rg[idx550]

                    # save fitted parameter values safely
                    try:
                        rho_east[i] = res.params['rho_s'].value
                        rho_d_east[i] = res.params['rho_d'].value
                        rho_s_east[i] = res.params['rho_s'].value
                    except Exception:
                        pass

                    # save full Result object for later inspection
                    res_east_list[i] = res

                else:  # WEST
                    Li_west[i, :] = Li_interp
                    Lt_west[i, :] = Lt_interp
                    Rrs_west[i, :] = Rrs_obs
                    Rrs_mod_west[i, :] = Rrs_mod
                    Rg_west[i, :] = Rg

                    resid_rmse_west[i] = rmse_val
                    resid550_west[i] = resid550_val
                    Rg550_west[i] = Rg[idx550]

                    try:
                        rho_west[i] = res.params['rho_s'].value
                        rho_d_west[i] = res.params['rho_d'].value
                        rho_s_west[i] = res.params['rho_s'].value
                    except Exception:
                        pass

                    res_west_list[i] = res


end_time = time.time()
elapsed_time = end_time - start_time  # in seconds
```

## Save results in netCDF with xarray

```python
# --- Create xarray Dataset and save to NetCDF ---

# Define coordinates
dates = unique_times  # datetime objects
dims = ("time", "wavelength")
coords = {"time": dates, "wavelength": wl}

# Create Dataset
ds = xr.Dataset(
    {
        "Es": (dims, Es_mat, {"long_name": "Downwelling irradiance Es", "units": "mW m^-2 nm^-1"}),
        "Li_east": (dims, Li_east, {"long_name": "Sky radiance East", "units": "mW m^-2 nm^-1 sr^-1"}),
        "Lt_east": (dims, Lt_east, {"long_name": "Surface radiance East", "units": "mW m^-2 nm^-1 sr^-1"}),
        "Rrs_east": (dims, Rrs_east, {"long_name": "Observed remote sensing reflectance East", "units": "sr^-1"}),
        "Rrs_mod_east": (dims, Rrs_mod_east, {"long_name": "Modeled remote sensing reflectance East", "units": "sr^-1"}),
        "Rg_east": (dims, Rg_east, {"long_name": "Surface-reflected reflectance East (R_g)", "units": "sr^-1"}),
        "Li_west": (dims, Li_west, {"long_name": "Sky radiance West", "units": "mW m^-2 nm^-1 sr^-1"}),
        "Lt_west": (dims, Lt_west, {"long_name": "Surface radiance West", "units": "mW m^-2 nm^-1 sr^-1"}),
        "Rrs_west": (dims, Rrs_west, {"long_name": "Observed remote sensing reflectance West", "units": "sr^-1"}),
        "Rrs_mod_west": (dims, Rrs_mod_west, {"long_name": "Modeled remote sensing reflectance West", "units": "sr^-1"}),
        "Rg_west": (dims, Rg_west, {"long_name": "Surface-reflected reflectance West (R_g)", "units": "sr^-1"})
    },
    coords=coords,
    attrs={
        "title": "Rrs time-series at the",
        "history": f"Created {datetime.now().isoformat()}",
        "references": "Groetsch et al. (2017), Pitarch (2026, in prep.)"
    }
)
```

## A few representative plots

```python
# Plot Es at 550 nm
plt.figure()
plt.plot(unique_times, Es_mat[:, wl == 550], 'k.-')
plt.ylabel('E_s(550)')
plt.gcf().autofmt_xdate()
plt.grid(True)
plt.show()

# Plot Rrs for East and West at 550 nm over time (existing)
plt.figure()
plt.plot(unique_times, Rrs_east[:, wl == 550], 'b.-', label='East')
plt.plot(unique_times, Rrs_west[:, wl == 550], 'r.-', label='West')
plt.ylabel('Rrs(550) (sr^-1)')
plt.gcf().autofmt_xdate()
plt.legend()
plt.grid(True)
plt.title('Time series of Rrs at 550 nm')
plt.show()

# === Additional diagnostics requested: residuals and fitted coefficients ===
# (The arrays were filled inside the fitting loop.)

# Plot diagnostics in a compact grid: 3 rows x 2 columns
fig, axs = plt.subplots(3, 2, figsize=(14, 10), sharex=True)
# Row 0, col 0: RMSE residuals
axs[0, 0].plot(unique_times, resid_rmse_east, 'b.-', label='East')
axs[0, 0].plot(unique_times, resid_rmse_west, 'r.-', label='West')
axs[0, 0].set_ylabel('RMSE (Lt/Es residual)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Row 0, col 1: Residual at 550 nm
axs[0, 1].plot(unique_times, resid550_east, 'b.-', label='East')
axs[0, 1].plot(unique_times, resid550_west, 'r.-', label='West')
axs[0, 1].set_ylabel('Residual @550 (Lt/Es)')
axs[0, 1].grid(True)

# Row 1, col 0: rho_s
axs[1, 0].plot(unique_times, rho_east, 'b.-', label='rho East')
axs[1, 0].plot(unique_times, rho_west, 'r.-', label='rho West')
axs[1, 0].set_ylabel('rho')
axs[1, 0].grid(True)
axs[1, 0].legend()

# Row 1, col 1: rho_dd
axs[1, 1].plot(unique_times, rho_d_east, 'b.-', label='rho_d East')
axs[1, 1].plot(unique_times, rho_d_west, 'r.-', label='rho_d West')
axs[1, 1].set_ylabel('rho_d')
axs[1, 1].grid(True)
axs[1, 1].legend()

# Row 2, col 0: rho_ds
axs[2, 0].plot(unique_times, rho_s_east, 'b.-', label='rho_s East')
axs[2, 0].plot(unique_times, rho_s_west, 'r.-', label='rho_s West')
axs[2, 0].set_ylabel('rho_s')
axs[2, 0].grid(True)
axs[2, 0].legend()

# Row 2, col 1: Rg @ 550 nm
axs[2, 1].plot(unique_times, Rg550_east, 'b.-', label='Rg@550 East')
axs[2, 1].plot(unique_times, Rg550_west, 'r.-', label='Rg@550 West')
axs[2, 1].set_ylabel('R_g (sr^-1) @550')
axs[2, 1].grid(True)
axs[2, 1].legend()

# X-axis formatting
for ax in axs[-1, :]:
    ax.xaxis.set_tick_params(rotation=30)
fig.tight_layout()
plt.show()
```
